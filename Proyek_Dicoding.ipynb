# Analisis Lanjutan
RFM Analysis

```python
import datetime as dt

now = dt.datetime(2018, 10, 17)

all_data['order_purchase_timestamp'] = pd.to_datetime(all_data['order_purchase_timestamp'])

# Group by 'customer_id' and calculate Recency, Frequency, and Monetary
recency = (now - all_data.groupby('customer_id')['order_purchase_timestamp'].max()).dt.days
frequency = all_data.groupby('customer_id')['order_id'].count()
monetary = all_data.groupby('customer_id')['price'].sum()
```

```python
rfm = pd.DataFrame({
    'customer_id': recency.index,
    'Recency': recency.values,
    'Frequency': frequency.values,
    'Monetary': monetary.values
})

# Calculate Percentiles
rfm_percentiles = rfm[['Recency', 'Frequency', 'Monetary']].rank(pct=True)

# RFM Segmentation
rfm['RFM_Score'] = rfm_percentiles['Recency'] * 100 + rfm_percentiles['Frequency'] * 10 + rfm_percentiles['Monetary']
rfm['RFM_Score'] = rfm['RFM_Score'].clip(0, 100)
rfm_segment = pd.cut(rfm['RFM_Score'], bins=[0, 33, 66, 100], labels=['Low', 'Mid', 'High'])
rfm['Segment'] = rfm_segment
```

```python
# Visualize RFM Segments
sns.scatterplot(x='Recency', y='Frequency', hue='Segment', data=rfm)
plt.title('RFM Segmentation')
plt.xlabel('Recency (Days Since Last Purchase)')
plt.ylabel('Frequency (Number of Purchases)')
plt.show()
```

-Sumbu horizontal (X-axis) menggambarkan nilai Recency, yaitu jumlah hari sejak pelanggan terakhir kali melakukan pembelian. Semakin rendah angkanya, semakin baru pelanggan tersebut melakukan transaksi. -Sementara itu, sumbu vertikal (Y-axis) merepresentasikan nilai Frequency, atau seberapa sering pelanggan melakukan pembelian. Nilai yang lebih tinggi berarti pelanggan lebih sering berbelanja. Dari visualisasi tersebut, terlihat bahwa sebagian besar pelanggan memiliki frekuensi pembelian yang rendah, yakni kurang dari 10 kali. Selain itu, banyak pelanggan yang sudah cukup lama tidak melakukan pembelian, yang terlihat dari segmen high yang lebih dominan dibandingkan segmen lainnya.

# Keterangan dari kolom-kolom analisis RFM


*   Recency: Menggambarkan berapa lama waktu yang telah berlalu sejak pelanggan terakhir kali melakukan pembelian. Semakin rendah nilai Recency, semakin baru transaksi yang dilakukan oleh pelanggan.

*   Frequency: Mengindikasikan seberapa sering pelanggan melakukan pembelian dalam jangka waktu tertentu. Pelanggan dengan nilai Frequency yang tinggi cenderung melakukan pembelian secara rutin.

*   Monetary: Mengukur total nilai pembelian atau pengeluaran pelanggan. Nilai Monetary yang tinggi menunjukkan bahwa pelanggan tersebut membelanjakan lebih banyak uang.

*   RFM_Score: Merupakan skor gabungan dari Recency, Frequency, dan Monetary. Skor RFM yang tinggi menunjukkan pelanggan yang aktif, sering berbelanja, dan memberikan kontribusi besar terhadap pendapatan perusahaan.

*   Segment: Segmen "High" mencakup pelanggan yang baru-baru ini sering berbelanja dan menghabiskan banyak uang, sedangkan segmen "Low" mencakup pelanggan yang jarang berbelanja dan memiliki pengeluaran rendah. Segmen "Mid" mencakup pelanggan dengan aktivitas belanja yang moderat dalam hal Recency, Frequency, dan Monetary.
